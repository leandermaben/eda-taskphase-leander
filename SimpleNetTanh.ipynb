```python
class Neural(nn.Module):
    def __init__(self,input_data,h,a,b):
        super(Neural,self).__init__()
        self.linear1=nn.Linear(input_data,h)
        nn.init.xavier_uniform_(self.linear1.weight)
        self.linear2=nn.Linear(h,a)
        nn.init.xavier_uniform_(self.linear2.weight)
        self.linear3=nn.Linear(a,b)
        nn.init.xavier_uniform_(self.linear3.weight)
        self.linear4=nn.Linear(b,1)
        #self.b1=nn.BatchNorm1d(a)
        #self.b2=nn.BatchNorm1d(b)
        #self.b3=nn.BatchNorm1d(c)
    def forward(self,x):
        x=torch.tanh(self.linear1(x))
        x=torch.tanh(self.linear2(x))
        x=torch.tanh(self.linear3(x))
        x=torch.sigmoid(self.linear4(x))
        return x
```

```python
class Data(Dataset):
    def __init__(self,df,transform=None):
        self.transform=transform
        self.data=df
        self.len=df.shape[0]
        
    def __len__(self):
        return self.len
    
    def __getitem__(self,ind):
        return torch.tensor(self.data.iloc[ind][0:20]).view(-1,20),torch.tensor(df.iloc[ind][20]).view(-1,1)
```

```python
dataset=Data(df)
```

```python
dataloader=DataLoader(dataset=dataset,batch_size=5)
```

```python
model=Neural(20,7,10,7)
criterion=nn.BCELoss()
cost=[]
optimizer=optim.Adam(model.parameters(),lr=0.001)
```

```python
for epoch in range(300):
    total=0
    for x,y in dataloader:
        yhat=model(x).view(-1,1)
        loss=criterion(yhat,y)
        total+=loss
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(epoch)
    print(loss)
    cost.append(loss)
```
